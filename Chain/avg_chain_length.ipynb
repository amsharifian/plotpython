```python
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib
from pylab import *
```

```python
# Setting plot style
# Print available styles

# print(plt.style.available)

matplotlib.style.use('seaborn-ticks')
```

```python
# Reading data from CSV file and putting workload name as index column
data = pd.read_csv('/home/amirali/git/plotpython/Data/avg_chain_length.csv', index_col='workload')
data.columns.values
```

```python
# Selecting columns 
n = data.loc[:,['ilp','merge']]
n.columns = ['Max ILP', 'Max size']
```

```python
# Ploting the bar chart
ax = n.plot(kind='barh', width=0.8, fontsize=15)

# Adding grid lines
ax.xaxis.grid()

# Setting X and Y labels
ax.set_xlabel("")
ax.set_ylabel("Average Chain Length", fontsize=20)

# Making plot tight
plt.tight_layout()
plt.show()
```
