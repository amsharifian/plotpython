```python
# Reading data from CSV file and putting workload name as index column
data = pd.read_csv('/home/amirali/git/plotpython/Data/chain_size_histo.csv', index_col='workload')
data.columns.values
```

```python
# Selecting columns 
n = data.loc[:,['1','2','3','4','5']]
n.columns = ['1','2','3','4','5+']
```

```python
# Ploting the bar chart
ax = n.plot(kind='bar', stacked=True, width=0.75, fontsize=15)

# Adding grid lines
ax.yaxis.grid()

# Setting X and Y labels
ax.set_xlabel("")
ax.set_ylabel("Number of Ops", fontsize=20)

# Making plot tight
plt.tight_layout()
plt.show()

```
