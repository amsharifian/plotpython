```python
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib
from pylab import *
from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
# Setting plot style
# Print available styles

# print(plt.style.available)


# matplotlib.style.use('ggplot')
matplotlib.style.use('seaborn-colorblind')
```

```python
# Reading data from CSV file and putting workload name as index column
data = pd.read_csv('/home/amirali/git/plotpython/Data/chain_size_histo.csv', index_col='workload')
data.columns.values
```

```python
# Selecting columns 
n = data.loc[:,['1','2','3','4','5']]
n.columns = ['1 Op','2 Ops','3 Ops','4 Ops','5+ Ops']
```

```python
#colors
colors=reversed(['#de2d26','#cccccc','#969696','#636363','#252525'])

# Ploting the bar chart
ax = n.plot(kind='bar', stacked=True, width=0.75,color=colors, fontsize=15)


ax.set_ylim([0,100])


# Adding grid lines
ax.yaxis.grid()

# Setting X and Y labels
ax.set_xlabel("")
ax.set_ylabel("Percentage of Ops", fontsize=20)

ax.legend(loc='upper center',bbox_to_anchor=(0.5,1.20),ncol=5,fancybox=True)

# Making plot tight
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

```
